```python
%matplotlib tk
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_random_vectors(num_vecs: int, dim: int):
    rng = np.random.default_rng()
    vecs = rng.standard_normal((num_vecs, dim))
    return vecs
```

```python
vecs = get_random_vectors(20, 3)
vecs_2d = get_random_vectors(20, 2)
```

```python
def plot_quiver_3d(vecs):
    fig = plt.figure()
    ax = fig.add_subplot(projection='3d')
    origin = np.zeros_like(vecs)
    ax.set_xlim3d(-2,2)
    ax.set_ylim3d(-2,2)
    ax.set_zlim3d(-2,2)
    ax.quiver(origin[:, 0], origin[:, 1], origin[:, 2], vecs[:, 0], vecs[:, 1], vecs[:, 2])
```

```python
def plot_quiver_2d(vecs):
    fig = plt.figure()
    ax = fig.add_subplot()
    origin = np.zeros_like(vecs)
    ax.set_xlim(-2, 2)
    ax.set_ylim(-2, 2)
    ax.quiver(origin[:, 0], origin[:, 1], vecs[:, 0], vecs[:, 1])
```

```python
def plot_scatter_2d(vecs, title):
    fig = plt.figure()
    ax = fig.add_subplot()
    ax.scatter(vecs[:, 0], vecs[:, 1])
    ax.set_xlim(-2, 2)
    ax.set_ylim(-2, 2)
    ax.set_title(title)
```

```python
def layer_normalization(vecs):
    normalized_vecs = []
    for vec in vecs:
        m = vec.mean()
        s = vec.std()
        
        normalized_vec = (vec - m) / s
        normalized_vecs.append(normalized_vec)
        
    return np.asarray(normalized_vecs)
```

```python
def batch_normalization(vecs):
    mx = vecs[:, 0].mean()
    my = vecs[:, 1].mean()
    sx = vecs[:, 0].std()
    sy = vecs[:, 1].std()

    vecs[:, 0] = (vecs[:, 0] - mx)
    vecs[:, 1] = (vecs[:, 1] - my)

    return vecs
```

```python
plot_quiver_3d(layer_normalization(vecs))
```

```python
rand_vec_2d = get_random_vectors(100, 2)
moved_vec_2d = rand_vec_2d + 0.5
plot_scatter_2d(moved_vec_2d, 'plus 0.5')
layer_vec_2d = layer_normalization(rand_vec_2d)
plot_scatter_2d(layer_vec_2d, 'layer norm')
batch_vec_2d = batch_normalization(rand_vec_2d)
plot_scatter_2d(batch_vec_2d, 'batch norm')
```
